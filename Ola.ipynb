```python
from functools import reduce

class ToDoList():
    def __init__(self):
        self.tasks = {}
    
    def add_task(self, title, duration):
        task_id = len(self.tasks) + 1
        self.tasks[task_id] = {"Title": title, "Duration": duration}
    
    def remove_task(self, task_id):
        if task_id in self.tasks:
            del self.tasks[task_id]
        else:
            print("Please enter valid Input. Task don't Exist")

    def view_tasks(self):
        if not self.tasks:
            print("No tasks available to show")
        else:
            for task_id, task in self.tasks.items():
                print(f"{task_id}: {task['Title']} - {task['Duration']}")

    def view_task_durations(self):
        task_duration = [task["Duration"] for task in self.tasks.values()]
        return task_duration

    def get_long_tasks(self, min_duration):
        long_tasks = [task for task in self.tasks.values() if task["Duration"] >= min_duration]
        return long_tasks

    def get_total_task_duration(self):
        total_duration = sum([task["Duration"] for task in self.tasks.values()])
        return total_duration

def main():
    todolist = ToDoList()
    
    while True:
        print("\n1. Add Task\n2. Remove Task\n3. View Tasks\n4. View Tasks Duration\n5. View Long Tasks\n6. View Total Task Duration\n7. Exit")
        choice = int(input("Enter your choice: "))
        
        if choice == 1:
            title = input("Enter Task Title: ")
            duration = int(input("Enter Task Duration: "))
            todolist.add_task(title, duration)
        elif choice == 2:
            task_id = int(input("Enter Task ID to remove: "))
            todolist.remove_task(task_id)
        elif choice == 3:
            todolist.view_tasks()
        elif choice == 4:
            print(todolist.view_task_durations())
        elif choice == 5:
            min_duration = int(input("Enter minimum duration to get long tasks: "))
            print(todolist.get_long_tasks(min_duration))
        elif choice == 6:
            print(todolist.get_total_task_duration())
        elif choice == 7:
            break

if __name__ == "__main__":
    main()
```
